```python
def fill_terms_for_steps(df, step_months, lgd_col='LGD_RATE_WITHOUT_TIME', ccf_col='CCF_WITHOUT_TIME', fallback_lgd='LGD_value', fallback_ccf='CCF'):
    """
    Fill LGD and CCF columns for all time steps according to rules:
    1. If LGD_RATE_WITHOUT_TIME or CCF_WITHOUT_TIME is present, replicate for all steps.
    2. If missing, use fallback LGD_value and CCF, replicate for all steps.
    Only fill LGD_{i} or CCF_{i} if the column is missing or empty (all NA or blank) in df.
    If LGD_{i} or CCF_{i} does not exist, create it and fill with lgd_source or ccf_source.
    """
    n_steps = len(step_months)
    # LGD
    lgd_clean = pd.to_numeric(df[lgd_col], errors='coerce') if lgd_col in df.columns else pd.Series([np.nan]*len(df))
    lgd_source = lgd_clean.combine_first(df[fallback_lgd])
    for i in range(1, n_steps + 1):
        col_name = f'LGD_{i}'
        if col_name not in df.columns:
            df[col_name] = lgd_source
        else:
            df[col_name] = pd.to_numeric(df[col_name], errors='coerce').fillna(lgd_source)
    # CCF
    ccf_clean = pd.to_numeric(df[ccf_col], errors='coerce') if ccf_col in df.columns else pd.Series([np.nan]*len(df))
    ccf_source = ccf_clean.combine_first(df[fallback_ccf])
    for i in range(1, n_steps + 1):
        col_name = f'CCF_{i}'
        if col_name not in df.columns:
            df[col_name] = ccf_source
        else:
            df[col_name] = pd.to_numeric(df[col_name], errors='coerce').fillna(ccf_source)
    return df
```

```python
def extend_pd_columns(df, nb_steps_col="NB_TIME_STEPS", pd_prefix="PD_"):
    """
    Extends PD columns in the DataFrame if contracts require more steps than the template provides.
    Fills new PD columns with the value of the last available PD for each row.
    """
    # Find the maximum number of steps required by any contract
    max_steps = df[nb_steps_col].max()
    # Find existing PD columns
    existing_pd_cols = [col for col in df.columns if re.match(rf"^{pd_prefix}\d+$", col)]
    max_existing_steps = max([int(col.split('_')[-1]) for col in existing_pd_cols]) if existing_pd_cols else 0
    existing_pd_cols_sorted = sorted(existing_pd_cols, key=lambda x: int(x.split('_')[-1]))
    # Add missing PD columns and fill with last PD value per row
    for i in range(1, max_steps + 1):
        col_name = f"{pd_prefix}{i}"
        if col_name not in existing_pd_cols:
            # Find last available PD column for each row
            last_pd_col = existing_pd_cols_sorted[-1] if existing_pd_cols_sorted else None
            if last_pd_col:
                df[col_name] = df[last_pd_col]
            else:
                df[col_name] = np.nan
    # Reorder columns so PD columns are in order
    pd_cols_sorted = [f"{pd_prefix}{i}" for i in range(1, max_steps + 1)]
    other_cols = [col for col in df.columns if col not in pd_cols_sorted]
    df = df[other_cols + pd_cols_sorted]
    return df
```

```python
def linear_ead_amortization(df, provisioning_basis_col, maturity_col, step_months, rate_col):
    """
    Calculate EAD amortization by time steps with constant payment and decreasing capital using pandas only.
    If rate_col is missing or empty, set to 0 for calculation only (do not modify df[rate_col]).
    Updates df in-place with new columns EAD_1, EAD_2, ..., EAD_n
    """
    n_steps = len(step_months)
    ead_cols = [f"EAD_{i+1}" for i in range(n_steps)]
    # Initial capital
    df[ead_cols[0]] = df[provisioning_basis_col]
    # Monthly rate, handle missing/empty rate_col (do not modify df[rate_col])
    rate_series = df[rate_col].fillna(0).replace('', 0).astype(float)
    monthly_rate = rate_series / 12
    # Total months
    total_months = df[maturity_col]

    # Annual payment (constant)
    annuity = np.where(
        rate_series == 0,
        df[provisioning_basis_col] / (total_months / 12),
        df[provisioning_basis_col] * rate_series / (1 - (1 + rate_series) ** (-total_months / 12))
    )
    # For each time step
    for i in range(1, n_steps):
        if i == 1:
            months = step_months[i-1]
        else:
            months = step_months[i-1] - step_months[i-2]
        last_ead_col = ead_cols[i-1]
        # Interest for the period
        interest = df[last_ead_col] * ((1 + monthly_rate) ** months - 1)
        # Constant payment (linear amortization)
        payment = months * annuity/12
        # Amortized capital for this period
        amortized_capital = payment - interest
        # Next EAD
        df[ead_cols[i]] = (df[last_ead_col] - amortized_capital).clip(lower=0)
    return df
```

```python
def off_balance_ead(df, provisioning_basis_col, nb_steps_col, step_months):
    """
    Calculate EAD for off-balance sheet items by time steps.
    Args:
        df: DataFrame with contract data
        provisioning_basis_col: column name for initial EAD
        step_months: array/list of time step boundaries
    """
    nb_steps = df[nb_steps_col]

    for i in range(1, len(step_months) + 1):
        mask = nb_steps >= i
        df.loc[mask, f"EAD_{i}"] = df.loc[mask, provisioning_basis_col] * df.loc[mask, f"CCF_{i}"]
        df.loc[~mask, f"EAD_{i}"] = 0
        
    return df
```

```python
def discount_factor(df, as_of_date_col, exposure_end_date_col, annual_rate_col, step_months):
    # Start time step 1 = Date end of month (As of date)

    # Start time step i > 1 = Date end of month (As of date + step_months[i-1])

    # End time step i = min(Date end of month (As of date + step_months[i-1]), Exposure end date)

    # Duration days i = (Start time step i - As of date) + (End time step i - Start time step i)/2

    # Discount factor i = 1 / (1 + annual_rate_col) ** (Duration days i/365)

    as_of_date = pd.to_datetime(df[as_of_date_col], errors='coerce', dayfirst=True)
    exposure_end_date = pd.to_datetime(df[exposure_end_date_col], errors='coerce', dayfirst=True)
    annual_rate = df[annual_rate_col].fillna(0).replace('', 0).astype(float)
    n_steps = len(step_months)

    for i in range(1, n_steps + 1):
        # Start of time step
        if i == 1:
            start_date = as_of_date
        else:
            start_date = as_of_date + pd.DateOffset(months=step_months[i-2])
        # End of time step
        end_date = as_of_date + pd.DateOffset(months=step_months[i-1]) - pd.Timedelta(days=1)
        end_date = end_date.combine(exposure_end_date, min)
        # Duration in days
        duration_days = (start_date - as_of_date).dt.days + ((end_date - start_date).dt.days) / 2
        # Discount factor
        df[f'DISCOUNT_{i}'] = 1 / (1 + annual_rate) ** (duration_days / 365)
    return df
```

```python
def discount_factor_quarterly(df, annual_rate_col, step_months):
    annual_rate = df[annual_rate_col].fillna(0).replace('', 0).astype(float)
    n_steps = len(step_months) 

    for i in range(1, n_steps + 1):
        if step_months[i - 1] <= 3:
            df[f'DF_{i}'] = 1.0
        else:
            n_quarters = (step_months[i - 1]) / 3
            df[f"DF_{i}"] = 1/(1 + annual_rate/4) ** n_quarters
    return df
```

```python
def apply_ead_amortization(df, step_months, rate_col="ANNUAL_RATE"):
    """
    Apply EAD amortization by time steps according to 'Amortization type' column, with default for empty values.
    Uses linear_ead_amortization for 'linear', constant_amortization for 'constant' and default.
    """
    mask_linear = df["AMORTIZATION_TYPE"] == "M-LINEAR"
    mask_constant = df["AMORTIZATION_TYPE"] == "I-FINE"
    mask_default = df["AMORTIZATION_TYPE"].isna() | (df["AMORTIZATION_TYPE"] == "")

    df_linear = df[mask_linear]
    df_constant = df[mask_constant]
    df_default = df[mask_default]

    # Apply functions
    df_linear_ead = amortization_ead.constant_amortization(
        df_constant,
        "PROVISIONING_BASIS",
        "RESIDUAL_MATURITY_MONTHS",
        step_months
    )

    df_constant_ead = amortization_ead.constant_amortization(
        df_constant,
        "PROVISIONING_BASIS",
        "RESIDUAL_MATURITY_MONTHS",
        step_months
    )

    # Default case: use constant amortization
    df_default_ead = amortization_ead.constant_amortization(
        df_default,
        "PROVISIONING_BASIS",
        "RESIDUAL_MATURITY_MONTHS",
        step_months
    )

    # Concatenate results
    df_ead = pd.concat([df_linear_ead, df_constant_ead, df_default_ead], axis=0).sort_index()
    return df_ead
```

```python
def pd_interpolation(df, step_months, method="linear", pd_prefix="PD_"):
    """
    Vectorized PD interpolation for the last time step using a common step_months array.
    Handles nb_steps == 1 case specifically.
    """
    nb_steps = df["NB_TIME_STEPS"].astype(int).to_numpy()
    residual_maturity = df["RESIDUAL_MATURITY_MONTHS"].to_numpy()
    as_of_date = pd.to_datetime(df["AS_OF_DATE"], errors="coerce", dayfirst=True)
    step_months_arr = np.array(step_months)
    month_offset_last = step_months_arr[nb_steps - 1]
    # Add months using numpy (month precision)
    as_of_date_month = as_of_date.values.astype('datetime64[M]')
    start_date_last_step = as_of_date_month + month_offset_last.astype('timedelta64[M]')
    # Adjust to last day of month
    start_date_last_step = pd.to_datetime(start_date_last_step) + pd.offsets.MonthEnd(0)
    # If you need end_date_last_step, use exposure_end_date or similar logic
    exposure_end_date = pd.to_datetime(df["EXPOSURE_END_DATE"], errors="coerce", dayfirst=True)
    end_date_last_step = np.minimum(start_date_last_step.values, exposure_end_date.values)
    # Duration in days
    duration_days = (end_date_last_step - start_date_last_step.values).astype("timedelta64[D]").astype(int)
    # ...existing code for PD interpolation...
    full_duration = np.where(nb_steps == 1, step_months[0], step_months[nb_steps - 1] - step_months[nb_steps - 2])
    real_duration = np.where(nb_steps == 1, residual_maturity, residual_maturity - step_months[nb_steps - 2])
    weight = real_duration / full_duration
    pd_cols = [f"{pd_prefix}{i+1}" for i in range(len(step_months))]
    PD_matrix = df[pd_cols].to_numpy()
    PD_last = PD_matrix[np.arange(len(df)), nb_steps - 1]
    if method == "linear":
        PD_last_interp = PD_last * weight
    elif method == "survival":
        PD_cum_prev = PD_matrix[np.arange(len(df)), :nb_steps - 1].sum(axis=1)
        PD_cum_last = PD_matrix[np.arange(len(df)), :nb_steps].sum(axis=1)
        PD_last_interp = 1 - (1 - PD_cum_prev) ** (1 - weight) * (1 - PD_cum_last) ** weight
    else:
        raise ValueError("Unknown interpolation method")
    for idx, n in enumerate(nb_steps):
        df.at[df.index[idx], f"{pd_prefix}{n}"] = PD_last_interp[idx]
    return df
```

```python
def get_matrix_prefix(df, prefix, round=False, round_nb:int=7):
    """
    Extracts columns with a given prefix and returns them as a NumPy matrix.
    Optionally rounds the values to a specified number of decimal places.
    """
    cols_sorted = []
    cols = [col for col in df.columns if re.match(rf"^{prefix}\d+$", col)]
    cols_sorted = sorted(cols, key=lambda x: int(x.split('_')[-1]))
    cols_sorted_val = df[cols_sorted].values
    if round:
        cols_sorted_val = np.round(cols_sorted_val, round_nb)
    return cols_sorted_val
```

```python
def calcul_ecl(df, step_months, list_scenarios):
    """
    Calculate ECL for each contract in the DataFrame for all time steps, 1 year, and lifetime.
    """
    all_steps = len(step_months)
    

    for scenario in list_scenarios:
        # Get PD terms
        scen_df = get_terms.get_terms_from_template(df, cst.PD_SHEET_MAPPING_CONFIG, 
                                            key, a.ecl_operation_data.template_data, prefix="PD_", scenario=scenario)
        scen_df = get_terms.extend_terms_columns(scen_df, "NB_TIME_STEPS", "PD_")
        scen_df = pd_interpolation(scen_df, step_months, method="linear", pd_prefix="PD_")

        # Get LGD terms
        scen_df = get_terms.get_terms_from_template(scen_df, cst.LGD_SHEET_MAPPING_CONFIG, key, 
                                            a.ecl_operation_data.template_data, prefix="LGD_", scenario=scenario)
        scen_df = get_terms.extend_terms_columns(scen_df, "NB_TIME_STEPS", "LGD_")

        # Get CCF terms
        scen_df = get_terms.get_terms_from_template(scen_df, cst.CCF_SHEET_MAPPING_CONFIG, key, 
                                            a.ecl_operation_data.template_data, prefix="CCF_", scenario=scenario)
        scen_df = get_terms.extend_terms_columns(scen_df, "NB_TIME_STEPS", "CCF_")
        scen_df = get_terms.fill_terms_for_lgd_ccf(scen_df, step_months, 
                                                   lgd_col='LGD_RATE_WITHOUT_TIME', ccf_col='CCF_WITHOUT_TIME', 
                                                   fallback_lgd='LGD_value', fallback_ccf='CCF')
        
        # Calculate EAD
        scen_df = apply_ead_amortization(scen_df, step_months, rate_col="CONTRACTUAL_CLIENT_RATE")

        # Get matrix for EAD, PD, LGD, DISCOUNT
        EAD_matrix = get_matrix_prefix(scen_df, "EAD_")
        PD_matrix = get_matrix_prefix(scen_df, "PD_", round=True, round_nb=7)
        LGD_matrix = get_matrix_prefix(scen_df, "LGD_", round=True, round_nb=7)
        CCF_matrix = get_matrix_prefix(scen_df, "CCF_", round=True, round_nb=7)
        DF_matrix = get_matrix_prefix(scen_df, "DISCOUNT_")

        ECL = EAD_matrix * PD_matrix * LGD_matrix * DF_matrix

        ecl_1y_cols = [i for i in range(all_steps) if step_months[i] <= 12]
        ECL_1Y = ECL[:, ecl_1y_cols].sum(axis=1)
        ECL_LT = ECL.sum(axis=1)

        # Add ECL columns to DataFrame
        ECL_1Y_df = pd.DataFrame(ECL_1Y, columns=[f"ECL_1Y_{scenario}"], index=scen_df.index)
        ECL_LT_df = pd.DataFrame(ECL_LT, columns=[f"ECL_LT_{scenario}"], index=scen_df.index)

        df = pd.concat([df, ECL_1Y_df, ECL_LT_df], axis=1)

    return df
```

```python
# Vectorized mapping of simulation data to LGD template using merge_asof
import pandas as pd

# Load LGD template and simulation data
lgd_template = pd.read_excel("LGD_template.xlsx")  # columns: PRODUCT, MATURITY_MAX, ...
simu_df = pd.read_csv("simulation_data.csv")      # columns: PRODUCT, RESIDUAL_MATURITY, ...

# Sort both DataFrames for merge_asof
lgd_template_sorted = lgd_template.sort_values(["PRODUCT", "MATURITY_MAX"])
simu_df_sorted = simu_df.sort_values(["PRODUCT", "RESIDUAL_MATURITY"])

# merge_asof requires grouping by PRODUCT, so we merge for each PRODUCT separately
result_list = []
for product in simu_df_sorted["PRODUCT"].unique():
    simu_sub = simu_df_sorted[simu_df_sorted["PRODUCT"] == product]
    lgd_sub = lgd_template_sorted[lgd_template_sorted["PRODUCT"] == product]
    merged = pd.merge_asof(
        simu_sub.sort_values("RESIDUAL_MATURITY"),
        lgd_sub.sort_values("MATURITY_MAX"),
        left_on="RESIDUAL_MATURITY",
        right_on="MATURITY_MAX",
        direction="backward"
    )
    result_list.append(merged)

# Concatenate all products
simu_lgd_mapped = pd.concat(result_list, ignore_index=True)

# Now simu_lgd_mapped contains all simulation rows with the closest LGD template row (not exceeding RESIDUAL_MATURITY)
simu_lgd_mapped.head()
```

```python
# Fill unmatched rows (where template columns are NaN) with max(MATURITY_MAX) for each PRODUCT
for product in simu_lgd_mapped["PRODUCT"].unique():
    mask = (simu_lgd_mapped["PRODUCT"] == product) & (simu_lgd_mapped["MATURITY_MAX"].isna())
    max_maturity = lgd_template[lgd_template["PRODUCT"] == product]["MATURITY_MAX"].max()
    # Fill template columns for unmatched rows
    for col in lgd_template.columns:
        if col == "PRODUCT":
            simu_lgd_mapped.loc[mask, col] = product
        elif col == "MATURITY_MAX":
            simu_lgd_mapped.loc[mask, col] = max_maturity
        else:
            # Fill with value from row with max(MATURITY_MAX) for this product
            value = lgd_template[(lgd_template["PRODUCT"] == product) & (lgd_template["MATURITY_MAX"] == max_maturity)][col].values
            if len(value) > 0:
                simu_lgd_mapped.loc[mask, col] = value[0]

simu_lgd_mapped.head()
```

```python
# Map RESIDUAL_MATURITY_MONTHS to closest (not exceeding) MATURITY_MAX by PRODUCT, then join on PRODUCT and new maturity_max_lgd column
import numpy as np

# Sort template for efficient mapping
lgd_template_sorted = lgd_template.sort_values(["PRODUCT", "MATURITY_MAX"])

# Function to map each row's RESIDUAL_MATURITY_MONTHS to closest MATURITY_MAX for its PRODUCT
def map_maturity_max(row, template):
    product = row["PRODUCT"]
    maturity = row["RESIDUAL_MATURITY_MONTHS"]
    candidates = template[(template["PRODUCT"] == product) & (template["MATURITY_MAX"] <= maturity)]
    if not candidates.empty:
        return candidates["MATURITY_MAX"].max()
    else:
        # If no candidate, use min MATURITY_MAX for that product (or np.nan)
        min_val = template[template["PRODUCT"] == product]["MATURITY_MAX"].min()
        return min_val if not np.isnan(min_val) else np.nan

# Create new column in simulation data
simu_df["maturity_max_lgd"] = simu_df.apply(lambda row: map_maturity_max(row, lgd_template_sorted), axis=1)

# Standard left join on PRODUCT and maturity_max_lgd
simu_lgd_mapped = simu_df.merge(lgd_template_sorted, how="left", left_on=["PRODUCT", "maturity_max_lgd"], right_on=["PRODUCT", "MATURITY_MAX"])

simu_lgd_mapped.head()
```

```python
def map_lgd_by_maturity(simu_df, lgd_template, product_col="PRODUCT", maturity_col="RESIDUAL_MATURITY_MONTHS", template_maturity_col="MATURITY_MAX"):
    """
    For each row in simu_df, map to the closest (not exceeding) MATURITY_MAX in lgd_template for the same PRODUCT.
    Adds a new column 'maturity_max_lgd' to simu_df and returns the merged DataFrame.
    """
    import numpy as np
    # Sort template for efficient mapping
    lgd_template_sorted = lgd_template.sort_values([product_col, template_maturity_col])
    # Function to map each row's maturity to closest template maturity
    def map_maturity_max(row):
        product = row[product_col]
        maturity = row[maturity_col]
        candidates = lgd_template_sorted[(lgd_template_sorted[product_col] == product) & (lgd_template_sorted[template_maturity_col] <= maturity)]
        if not candidates.empty:
            return candidates[template_maturity_col].max()
        else:
            min_val = lgd_template_sorted[lgd_template_sorted[product_col] == product][template_maturity_col].min()
            return min_val if not np.isnan(min_val) else np.nan
    # Create new column in simulation data
    simu_df = simu_df.copy()
    simu_df["maturity_max_lgd"] = simu_df.apply(map_maturity_max, axis=1)
    # Standard left join on PRODUCT and maturity_max_lgd
    merged_df = simu_df.merge(lgd_template_sorted, how="left", left_on=[product_col, "maturity_max_lgd"], right_on=[product_col, template_maturity_col])
    return merged_df

# Example usage:
# mapped_df = map_lgd_by_maturity(simu_df, lgd_template)
# mapped_df.head()
```
